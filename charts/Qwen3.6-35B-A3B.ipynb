```python
import plotly.graph_objects

oq_data = [
    {"ram": 15.1, "kld": 1.794922, "label": "oQ2", "pos": "top right"},
    {"ram": 18.75, "kld": 0.636719, "label": "oQ3", "pos": "top right"},
    {"ram": 19.97, "kld": 0.55957, "label": "oQ3.5", "pos": "top right"},
    {"ram": 22.7, "kld": 0.116943, "label": "oQ4", "pos": "top right"},
    {"ram": 26.55, "kld": 0.037903, "label": "oQ5", "pos": "top right"},
    {"ram": 30.3, "kld": 0.025452, "label": "oQ6", "pos": "top right"},
    {"ram": 38.08, "kld": 0.009682, "label": "oQ8", "pos": "top right"},
]

q_data = [
    {"ram": 13.89, "kld": 4.6875, "label": "Q2", "pos": "top right"},
    {"ram": 17.93, "kld": 1.103516, "label": "Q3", "pos": "top right"},
    {"ram": 20.95, "kld": 0.383822, "label": "MXFP4", "pos": "top right"},
    {"ram": 21.96, "kld": 0.211914, "label": "Q4", "pos": "top right"},
    {"ram": 25.96, "kld": 0.111914, "label": "Q5", "pos": "top right"},
    {"ram": 30.02, "kld": 0.026718, "label": "Q6", "pos": "top left"},
    {"ram": 37.08, "kld": 0.130360, "label": "MXFP8", "pos": "top right"},
    {"ram": 38.09, "kld": 0.011803, "label": "Q8", "pos": "top left", "x": -0.1},
]

x_min, x_max = 14, 40
y_min, y_max = 0, 2

def extract(data):
    ram = [d["ram"] for d in data]
    kld = [d["kld"] for d in data]
    labels = [d["label"] for d in data]
    textpos = [d["pos"] for d in data]
    x_text = [d["ram"] + d.get("x", 0.0) for d in data]
    y_text = [d["kld"] + d.get("y", 0.02) for d in data]
    return ram, kld, labels, textpos, x_text, y_text

ram_oq, kld_oq, labels_oq, pos_oq, x_oq_text, y_oq_text = extract(oq_data)
ram_q,  kld_q,  labels_q,  pos_q,  x_q_text,  y_q_text  = extract(q_data)

fig = plotly.graph_objects.Figure()

scatter_markers_oq = plotly.graph_objects.Scatter(
    x=ram_oq,
    y=kld_oq,
    mode='markers',
    name='oQ',
    marker=dict(color='#1f77b4', size=8)
)
fig.add_trace(scatter_markers_oq)

scatter_markers_q = plotly.graph_objects.Scatter(
    x=ram_q,
    y=kld_q,
    mode='markers',
    name='Q',
    marker=dict(color='black', size=8)
)
fig.add_trace(scatter_markers_q)

scatter_labels_oq = plotly.graph_objects.Scatter(
    x=x_oq_text,
    y=y_oq_text,
    mode='text',
    name='oQ_labels',
    text=labels_oq,
    textposition=pos_oq,
    textfont=dict(color='#1f77b4', size=14),
    showlegend=False
)
fig.add_trace(scatter_labels_oq)

scatter_labels_q = plotly.graph_objects.Scatter(
    x=x_q_text,
    y=y_q_text,
    mode='text',
    name='Q_labels',
    text=labels_q,
    textposition=pos_q,
    textfont=dict(color='black', size=14),
    showlegend=False
)
fig.add_trace(scatter_labels_q)

height = 800
top_margin = 60
x_range = x_max - x_min
y_range = y_max - y_min
dx_minor = 1
dy_minor = 0.1
width = (height - top_margin) * (x_range / dx_minor) / (y_range / dy_minor)

fig.update_layout(
    title="Qwen3.6-35B-A3B",
    title_x=0.5,
    width=width,
    height=height,
    showlegend=False,
    font=dict(color="black"),
    xaxis_title="RAM (GiB)",
    yaxis_title="KL Divergence (nats)",
    yaxis=dict(
        range=[y_min, y_max],
        showgrid=True,
        gridcolor='lightgray',
        ticklabelposition="outside",
        ticklabelstandoff=10,
        minor=dict(
            showgrid=True,
            dtick=dy_minor,
            gridcolor='#f5f5f5',
            gridwidth=0.5
        ),
    ),
    xaxis=dict(
        range=[x_min, x_max],
        showgrid=True,
        gridcolor='lightgray',
        ticklabelposition="outside",
        ticklabelstandoff=10,
        minor=dict(
            showgrid=True,
            dtick=dx_minor,
            gridcolor='#f5f5f5',
            gridwidth=0.5
        ),
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(t=top_margin, b=0, l=0, r=0)
)

fig.show()
```
